```python
i=0
while i <=11000:
    
    Value1=np.random.randint(0,25)
    Value2=np.random.randint(0,25)
    Value3=np.random.randint(0,35)
    Value4=np.random.randint(0,35)
    Value5=np.random.randint(1,45)
    
    wlist =[Value1,Value2,Value3,Value4,Value5]
    wlist.sort() 

    a=wlist[0]
    b=wlist[1]
    c=wlist[2]
    d=wlist[3]
    e=wlist[4]+0.001
    
    
    Sum=(a+b+c+d+e)
    Mean=np.mean(wlist)
    alpha=random.choice([1,1])
    
    #####################################################################################################
    #####################################################################################################
    #Transfer
    Transfer=random.choice([0.1,0.2,0.3,0.4,0.45,0.49])
    TransferNr1=np.random.randint(1,10)
  
    
    if(TransferNr1==1):
        Value1tf=a+Transfer
        Value2tf=b-Transfer
        Value3tf=c
        Value4tf=d
        Value5tf=e
        
        Quantile1='b2a'
        TransferSum1=Transfer
        RankDiff1=1
        Difference1=abs(a-b)
        
    elif(TransferNr1==2):
        Value1tf=a+Transfer
        Value2tf=b
        Value3tf=c-Transfer
        Value4tf=d
        Value5tf=e
        
        Quantile1='c2a' 
        TransferSum1=Transfer
        RankDiff1=2
        Difference1=abs(a-c)
       
    elif(TransferNr1==3):
        Value1tf=a+Transfer
        Value2tf=b
        Value3tf=c
        Value4tf=d-Transfer
        Value5tf=e
        
        Quantile1='d2a'
        TransferSum1=Transfer
        RankDiff1=3
        Difference1=abs(a-d)
       
        
    elif(TransferNr1==4):
        Value1tf=a+Transfer
        Value2tf=b
        Value3tf=c
        Value4tf=d
        Value5tf=e-Transfer
       
        
        Quantile1='e2a'
        TransferSum1=Transfer
        RankDiff1=4
        Difference1=abs(a-e)
       
     
    ################################
    elif(TransferNr1==5):
        Value1tf=a
        Value2tf=b+Transfer
        Value3tf=c-Transfer
        Value4tf=d
        Value5tf=e
        
        Quantile1='c2b'
        TransferSum1=Transfer
        RankDiff1=1
        Difference1=abs(b-c)
        
                       
    elif(TransferNr1==6):
        Value1tf=a
        Value2tf=b+Transfer
        Value3tf=c
        Value4tf=d-Transfer
        Value5tf=e
        
        Quantile1='d2b'
        TransferSum1=Transfer
        RankDiff1=2
        Difference1=abs(b-d)
      
        
    elif(TransferNr1==7):
        Value1tf=a
        Value2tf=b+Transfer
        Value3tf=c
        Value4tf=d
        Value5tf=e-Transfer
        
        Quantile1='e2b'
        TransferSum1=Transfer
        RankDiff1=3
        Difference1=abs(b-e)
        
    ###################################
    
    elif(TransferNr1==8):
        Value1tf=a
        Value2tf=b
        Value3tf=c+Transfer
        Value4tf=d-Transfer
        Value5tf=e
       
        Quantile1='d2c'
        TransferSum1=Transfer
        RankDiff1=1
        Difference1=abs(c-d)
     
        
    elif(TransferNr1==9):
        Value1tf=a
        Value2tf=b
        Value3tf=c+Transfer
        Value4tf=d
        Value5tf=e-Transfer
        
        Quantile1='e2c'
        TransferSum1=Transfer
        RankDiff1=2
        Difference1=abs(c-e)
       
        
    ##########################
    elif(TransferNr1==10):
        Value1tf=a
        Value2tf=b
        Value3tf=c
        Value4tf=d+Transfer
        Value5tf=e-Transfer
    
        Quantile1='e2d'
        TransferSum1=Transfer
        RankDiff1=1
        Difference1=abs(d-e)
       
    #######################################################    
    wlisttf =[Value1tf,Value2tf,Value3tf,Value4tf,Value5tf]
    wlisttf.sort()
    
    atf=wlisttf[0]
    btf=wlisttf[1]
    ctf=wlisttf[2]
    dtf=wlisttf[3]
    etf=wlisttf[4]
    
   
    ##################################################################################################
    Sumtf=(atf+btf+ctf+dtf+etf)
    Meantf=Sumtf/5
    
    ###################################################################################################
    #####################################################################################################
    #Transfer 2
    TransferNr2=np.random.randint(1,10)
    ###############################
    
    if(TransferNr2==1):
        Value1tf2=a+Transfer
        Value2tf2=b-Transfer
        Value3tf2=c
        Value4tf2=d
        Value5tf2=e
        
        Quantile2='b2a'
        TransferSum2=Transfer
        RankDiff2=1
        Difference2=abs(a-b)
        
    elif(TransferNr2==2):
        Value1tf2=a+Transfer
        Value2tf2=b
        Value3tf2=c-Transfer
        Value4tf2=d
        Value5tf2=e
        
        Quantile2='c2a' 
        TransferSum2=Transfer
        RankDiff2=2
        Difference2=abs(a-c)
       
    elif(TransferNr2==3):
        Value1tf2=a+Transfer
        Value2tf2=b
        Value3tf2=c
        Value4tf2=d-Transfer
        Value5tf2=e
        
        Quantile2='d2a'
        TransferSum2=Transfer
        RankDiff2=3
        Difference2=abs(a-d)
       
        
    elif(TransferNr2==4):
        Value1tf2=a+Transfer
        Value2tf2=b
        Value3tf2=c
        Value4tf2=d
        Value5tf2=e-Transfer
       
        
        Quantile2='e2a'
        TransferSum2=Transfer
        RankDiff2=4
        Difference2=abs(a-e)
       
     
    ################################
    elif(TransferNr2==5):
        Value1tf2=a
        Value2tf2=b+Transfer
        Value3tf2=c-Transfer
        Value4tf2=d
        Value5tf2=e
        
        Quantile2='c2b'
        TransferSum2=Transfer
        RankDiff2=1
        Difference2=abs(b-c)
        
                       
    elif(TransferNr2==6):
        Value1tf2=a
        Value2tf2=b+Transfer
        Value3tf2=c
        Value4tf2=d-Transfer
        Value5tf2=e
        
        Quantile2='d2b'
        TransferSum2=Transfer
        RankDiff2=2
        Difference2=abs(b-d)
      
        
    elif(TransferNr2==7):
        Value1tf2=a
        Value2tf2=b+Transfer
        Value3tf2=c
        Value4tf2=d
        Value5tf2=e-Transfer
        
        Quantile2='e2b'
        TransferSum2=Transfer
        RankDiff2=3
        Difference2=abs(b-e)
        
    ###################################
    
    elif(TransferNr2==8):
        Value1tf2=a
        Value2tf2=b
        Value3tf2=c+Transfer
        Value4tf2=d-Transfer
        Value5tf2=e
       
        Quantile2='d2c'
        TransferSum2=Transfer
        RankDiff2=1
        Difference2=abs(c-d)
     
        
    elif(TransferNr2==9):
        Value1tf2=a
        Value2tf2=b
        Value3tf2=c+Transfer
        Value4tf2=d
        Value5tf2=e-Transfer
        
        Quantile2='e2c'
        TransferSum2=Transfer
        RankDiff2=2
        Difference2=abs(c-e)
       
        
    ##########################
    elif(TransferNr2==10):
        Value1tf2=a
        Value2tf2=b
        Value3tf2=c
        Value4tf2=d+Transfer
        Value5tf2=e-Transfer
    
        Quantile2='e2d'
        TransferSum2=Transfer
        RankDiff2=1
        Difference2=abs(d-e)
       
    #######################################################    
    wlisttf2 =[Value1tf2,Value2tf2,Value3tf2,Value4tf2,Value5tf2]
    wlisttf2.sort()
    
    atf2=wlisttf2[0]
    btf2=wlisttf2[1]
    ctf2=wlisttf2[2]
    dtf2=wlisttf2[3]
    etf2=wlisttf2[4]


    Sumtf2=abs(atf2+btf2+ctf2+dtf2+etf2)
    Meantf2=Sumtf2/5
    
    ######################################################################################################
    #####################################################################################################
    ####################################################################################################
    #ArcusDiff
    
    ArcusDiff12=abs((np.arctan2(a,b)*180/math.pi)-(np.arctan2(b,a)*180/math.pi)) 
    ArcusDiff13=abs((np.arctan2(a,c)*180/math.pi)-(np.arctan2(c,a)*180/math.pi)) 
    ArcusDiff14=abs((np.arctan2(a,d)*180/math.pi)-(np.arctan2(d,a)*180/math.pi))
    ArcusDiff15=abs((np.arctan2(a,e)*180/math.pi)-(np.arctan2(e,a)*180/math.pi))  
      
     
    ArcusDiff23=abs((np.arctan2(b,c)*180/math.pi)-(np.arctan2(c,b)*180/math.pi))  
    ArcusDiff24=abs((np.arctan2(b,d)*180/math.pi)-(np.arctan2(d,b)*180/math.pi))  
    ArcusDiff25=abs((np.arctan2(b,e)*180/math.pi)-(np.arctan2(e,b)*180/math.pi)) 
    
    ArcusDiff34=abs((np.arctan2(c,d)*180/math.pi)-(np.arctan2(d,c)*180/math.pi))
    ArcusDiff35=abs((np.arctan2(c,e)*180/math.pi)-(np.arctan2(e,c)*180/math.pi))
        
    ArcusDiff45=abs((np.arctan2(d,e)*180/math.pi)-(np.arctan2(e,d)*180/math.pi)) 
   

    ArcusDiffSum=(ArcusDiff12+ArcusDiff13+ArcusDiff14+ArcusDiff15+
              ArcusDiff23+ArcusDiff24+ArcusDiff25+
              ArcusDiff34+ArcusDiff35+
              ArcusDiff45)
    
    ArcusDiffMittelValue=ArcusDiffSum/10
    preArcusDiff=ArcusDiffMittelValue/90
    ArcusDiff=(preArcusDiff-(preArcusDiff*(1/5)))
    
    
    #########################################################################################################
    #ArcusDiff transfer

    
    ArcusDiff12tf=abs((np.arctan2(atf,btf)*180/math.pi)-(np.arctan2(btf,atf)*180/math.pi))
    ArcusDiff13tf=abs((np.arctan2(atf,ctf)*180/math.pi)-(np.arctan2(ctf,atf)*180/math.pi))
    ArcusDiff14tf=abs((np.arctan2(atf,dtf)*180/math.pi)-(np.arctan2(dtf,atf)*180/math.pi))
    ArcusDiff15tf=abs((np.arctan2(atf,etf)*180/math.pi)-(np.arctan2(etf,atf)*180/math.pi))
   
    ArcusDiff23tf=abs((np.arctan2(btf,ctf)*180/math.pi)-(np.arctan2(ctf,btf)*180/math.pi))
    ArcusDiff24tf=abs((np.arctan2(btf,dtf)*180/math.pi)-(np.arctan2(dtf,btf)*180/math.pi))
    ArcusDiff25tf=abs((np.arctan2(btf,etf)*180/math.pi)-(np.arctan2(etf,btf)*180/math.pi))
    
    ArcusDiff34tf=abs((np.arctan2(ctf,dtf)*180/math.pi)-(np.arctan2(dtf,ctf)*180/math.pi))
    ArcusDiff35tf=abs((np.arctan2(ctf,etf)*180/math.pi)-(np.arctan2(etf,ctf)*180/math.pi))
   
    ArcusDiff45tf=abs((np.arctan2(dtf,etf)*180/math.pi)-(np.arctan2(etf,dtf)*180/math.pi))
   

    ArcusDiffSumtf=(ArcusDiff12tf+ArcusDiff13tf+ArcusDiff14tf+ArcusDiff15tf+
                ArcusDiff23tf+ArcusDiff24tf+ArcusDiff25tf+
                ArcusDiff34tf+ArcusDiff35tf+
                ArcusDiff45tf)
    
    ArcusDiffMittelValuetf=ArcusDiffSumtf/10
    preArcusDifftf=ArcusDiffMittelValuetf/90
    ArcusDifftf=(preArcusDifftf-(preArcusDifftf*(1/5)))
     
    ArcusDiffd=100-((100/ArcusDiff)*ArcusDifftf) 
    
    ########################################################################
    #ArcusDiff transfer 2

    ArcusDiff12tf2=abs((np.arctan2(atf2,btf2)*180/math.pi)-(np.arctan2(btf2,atf2)*180/math.pi))
    ArcusDiff13tf2=abs((np.arctan2(atf2,ctf2)*180/math.pi)-(np.arctan2(ctf2,atf2)*180/math.pi))
    ArcusDiff14tf2=abs((np.arctan2(atf2,dtf2)*180/math.pi)-(np.arctan2(dtf2,atf2)*180/math.pi))
    ArcusDiff15tf2=abs((np.arctan2(atf2,etf2)*180/math.pi)-(np.arctan2(etf2,atf2)*180/math.pi))
   
    ArcusDiff23tf2=abs((np.arctan2(btf2,ctf2)*180/math.pi)-(np.arctan2(ctf2,btf2)*180/math.pi))
    ArcusDiff24tf2=abs((np.arctan2(btf2,dtf2)*180/math.pi)-(np.arctan2(dtf2,btf2)*180/math.pi))
    ArcusDiff25tf2=abs((np.arctan2(btf2,etf2)*180/math.pi)-(np.arctan2(etf2,btf2)*180/math.pi))
    
    ArcusDiff34tf2=abs((np.arctan2(ctf2,dtf2)*180/math.pi)-(np.arctan2(dtf2,ctf2)*180/math.pi))
    ArcusDiff35tf2=abs((np.arctan2(ctf2,etf2)*180/math.pi)-(np.arctan2(etf2,ctf2)*180/math.pi))
    
    ArcusDiff45tf2=abs((np.arctan2(dtf2,etf2)*180/math.pi)-(np.arctan2(etf2,dtf2)*180/math.pi))
    
    ArcusDiffSumtf2=(ArcusDiff12tf2+ArcusDiff13tf2+ArcusDiff14tf2+ArcusDiff15tf2+
                 ArcusDiff23tf2+ArcusDiff24tf2+ArcusDiff25tf2+
                 ArcusDiff34tf2+ArcusDiff35tf2+
                 ArcusDiff45tf2
                 )
    
    ArcusDiffMittelValuetf2=ArcusDiffSumtf2/10
    preArcusDifftf2=ArcusDiffMittelValuetf2/90
    ArcusDifftf2=(preArcusDifftf2-(preArcusDifftf2*(1/5)))
     
    ArcusDiffd2=100-((100/ArcusDiff)*ArcusDifftf2) 
    
    #############################################################################################
    ##########################################################################################
    #Vinci
    
    Gini12=abs(a-b)/(Mean*2)
    Gini13=abs(a-c)/(Mean*2)
    Gini14=abs(a-d)/(Mean*2)
    Gini15=abs(a-e)/(Mean*2)
   
    
    Gini23=abs(b-c)/(Mean*2)
    Gini24=abs(b-d)/(Mean*2)
    Gini25=abs(b-e)/(Mean*2)
    
    Gini34=abs(c-d)/(Mean*2)
    Gini35=abs(c-e)/(Mean*2)
   
    Gini45=abs(d-e)/(Mean*2)
    
    Vinci12= ArcusDiff12 * Gini12
    Vinci13= ArcusDiff13 * Gini13
    Vinci14= ArcusDiff14 * Gini14
    Vinci15= ArcusDiff15 * Gini15
   
    Vinci23= ArcusDiff23 * Gini23
    Vinci24= ArcusDiff24 * Gini24
    Vinci25= ArcusDiff25 * Gini25
    
    Vinci34= ArcusDiff34 * Gini34
    Vinci35= ArcusDiff35 * Gini35
   
    Vinci45= ArcusDiff45 * Gini45
   

    VinciSum=(Vinci12+Vinci13+Vinci14+Vinci15+
               Vinci23+Vinci24+Vinci25+
               Vinci34+Vinci35+
               Vinci45
               )
    
    VinciMittelValue=VinciSum/10
    preVinci=VinciMittelValue/90
    Vinci=(preVinci-(preVinci*(1/5)))
    
    #######################################################################################################
    #Vinci tf

    Gini12tf=abs(atf-btf)/(Mean*2)
    Gini13tf=abs(atf-ctf)/(Mean*2)
    Gini14tf=abs(atf-dtf)/(Mean*2)
    Gini15tf=abs(atf-etf)/(Mean*2)
   
    
    Gini23tf=abs(btf-ctf)/(Mean*2)
    Gini24tf=abs(btf-dtf)/(Mean*2)
    Gini25tf=abs(btf-etf)/(Mean*2)
    
    Gini34tf=abs(ctf-dtf)/(Mean*2)
    Gini35tf=abs(ctf-etf)/(Mean*2)
  
    
    Gini45tf=abs(dtf-etf)/(Mean*2)
   
    Vinci12tf= ArcusDiff12tf * Gini12tf
    Vinci13tf= ArcusDiff13tf * Gini13tf
    Vinci14tf= ArcusDiff14tf * Gini14tf
    Vinci15tf= ArcusDiff15tf * Gini15tf
   
    Vinci23tf= ArcusDiff23tf * Gini23tf
    Vinci24tf= ArcusDiff24tf * Gini24tf
    Vinci25tf= ArcusDiff25tf * Gini25tf
    
    Vinci34tf= ArcusDiff34tf * Gini34tf
    Vinci35tf= ArcusDiff35tf * Gini35tf
    Vinci45tf= ArcusDiff45tf * Gini45tf
   
    VinciSumtf=(Vinci12tf+Vinci13tf+Vinci14tf+Vinci15tf+
                 Vinci23tf+Vinci24tf+Vinci25tf+
                 Vinci34tf+Vinci35tf+
                 Vinci45tf)
    
    VinciMittelValuetf=VinciSumtf/10
    preVincitf=VinciMittelValuetf/90
    Vincitf=(preVincitf-(preVincitf*(1/5)))
    
    Vincid=100-((100/Vinci)*Vincitf)
    
    ###################################################################################################
    #Gini
    
    g1 = (a/(a+b+c+d+e))
    g2 = (b/(a+b+c+d+e))
    g3 = (c/(a+b+c+d+e))
    g4 = (d/(a+b+c+d+e))
    g5 = (e/(a+b+c+d+e))
   
   
    Sum = (g1+g2+g3+g4+g5)
    glist = [g1,g2,g3,g4,g5]
    glist.sort()
    
    v1=glist[0]
    v2=glist[1]
    v3=glist[2]
    v4=glist[3]
    v5=glist[4]
   
    
    c1=glist[0]
    c2=v1+glist[1]
    c3=v2+glist[2]
    c4=v3+glist[3]
    c5=v4+glist[4]
    
    freq=1/5
    f1=freq
    f2=f1+freq
    f3=f2+freq
    f4=f3+freq
    f5=f4+freq
   
    gi1=f1*v1
    gi2=(f1+f2)*v2
    gi3=(f2+f3)*v3
    gi4=(f3+f4)*v4
    gi5=(f4+f5)*v5
    
    giSum=(gi1+gi2+gi3+gi4+gi5)
    Gini=giSum-1
    
    ###############################################################################################
    #Gini tf
    
    g1tf = (atf/(atf+btf+ctf+dtf+etf))
    g2tf = (btf/(atf+btf+ctf+dtf+etf))
    g3tf = (ctf/(atf+btf+ctf+dtf+etf))
    g4tf = (dtf/(atf+btf+ctf+dtf+etf))
    g5tf = (etf/(atf+btf+ctf+dtf+etf))
    
   
    Sumtf = (g1tf+g2tf+g3tf+g4tf+g5tf)
    glisttf = [g1tf,g2tf,g3tf,g4tf,g5tf]
    glisttf.sort()
    
    v1tf=glisttf[0]
    v2tf=glisttf[1]
    v3tf=glisttf[2]
    v4tf=glisttf[3]
    v5tf=glisttf[4]
   
    
    c1tf=glisttf[0]
    c2tf=v1tf+glisttf[1]
    c3tf=v2tf+glisttf[2]
    c4tf=v3tf+glisttf[3]
    c5tf=v4tf+glisttf[4]
    
    freqtf=1/5
    f1tf=freqtf
    f2tf=f1tf+freqtf
    f3tf=f2tf+freqtf
    f4tf=f3tf+freqtf
    f5tf=f4tf+freqtf
  
  
    gi1tf=f1tf*v1tf
    gi2tf=(f1tf+f2tf)*v2tf
    gi3tf=(f2tf+f3tf)*v3tf
    gi4tf=(f3tf+f4tf)*v4tf
    gi5tf=(f4tf+f5tf)*v5tf
    
    giSumtf=(gi1tf+gi2tf+gi3tf+gi4tf+gi5tf)
    Ginitf=giSumtf-1
   
    Ginid=100-((100/Gini)*Ginitf)
    
    g1tf2 = (atf2/(atf2+btf2+ctf2+dtf2+etf2))
    g2tf2 = (btf2/(atf2+btf2+ctf2+dtf2+etf2))
    g3tf2 = (ctf2/(atf2+btf2+ctf2+dtf2+etf2))
    g4tf2 = (dtf2/(atf2+btf2+ctf2+dtf2+etf2))
    g5tf2 = (etf2/(atf2+btf2+ctf2+dtf2+etf2))
   
    Sumtf2 = (g1tf2+g2tf2+g3tf2+g4tf2+g5tf2)
    glisttf2 = [g1tf2,g2tf2,g3tf2,g4tf2,g5tf2]
    glisttf2.sort()
    
    #####################################################################
    #Gini tf2 
    
    Gini12tf2=abs(atf2-btf2)/(Mean*2)
    Gini13tf2=abs(atf2-ctf2)/(Mean*2)
    Gini14tf2=abs(atf2-dtf2)/(Mean*2)
    Gini15tf2=abs(atf2-etf2)/(Mean*2)

    Gini23tf2=abs(btf2-ctf2)/(Mean*2)
    Gini24tf2=abs(btf2-dtf2)/(Mean*2)
    Gini25tf2=abs(btf2-etf2)/(Mean*2)
    
    Gini34tf2=abs(ctf2-dtf2)/(Mean*2)
    Gini35tf2=abs(ctf2-etf2)/(Mean*2)

    Gini45tf2=abs(dtf2-etf2)/(Mean*2)
    
    
   
    SumGinitf2=(Gini12tf2+Gini13tf2+Gini14tf2+Gini15tf2+
                Gini23tf2+Gini24tf2+Gini25tf2+
                Gini34tf2+Gini35tf2+
                Gini45tf2
                )
    
    MeanGinitf2=SumGinitf2/10
    Ginitf2=(MeanGinitf2-(MeanGinitf2*(1/5)))
    
    Ginid2=100-((100/Gini)*Ginitf2)

   
    #######################################################################################
    #######################################################################################
    ####################################################################################
    #Vinci

    Vinci12=pow(ArcusDiff12,alpha) * Gini12
    Vinci13=pow(ArcusDiff13,alpha) * Gini13
    Vinci14=pow(ArcusDiff14,alpha) * Gini14
    Vinci15=pow(ArcusDiff15,alpha) * Gini15
    
    Vinci23=pow(ArcusDiff23,alpha) * Gini23
    Vinci24=pow(ArcusDiff24,alpha) * Gini24
    Vinci25=pow(ArcusDiff25,alpha) * Gini25

    Vinci34=pow(ArcusDiff34,alpha) * Gini34
    Vinci35=pow(ArcusDiff35,alpha) * Gini35

    Vinci45=pow(ArcusDiff45,alpha) * Gini45
    
    
    VinciSum=(Vinci12+Vinci13+Vinci14+Vinci15+
                Vinci23+Vinci24+Vinci25+
                Vinci34+Vinci35+
                Vinci45
                
                )
    
    VinciMittelValue=VinciSum/10
    preVinci=VinciMittelValue/pow(90,alpha)
    Vinci=preVinci-(preVinci*1/5)
    
    #########################################################################################################
    #Vinci tf
    
    Vinci12tf=pow(ArcusDiff12tf,alpha) * Gini12tf
    Vinci13tf=pow(ArcusDiff13tf,alpha) * Gini13tf
    Vinci14tf=pow(ArcusDiff14tf,alpha) * Gini14tf
    Vinci15tf=pow(ArcusDiff15tf,alpha) * Gini15tf
     
    Vinci23tf=pow(ArcusDiff23tf,alpha) * Gini23tf
    Vinci24tf=pow(ArcusDiff24tf,alpha) * Gini24tf
    Vinci25tf=pow(ArcusDiff25tf,alpha) * Gini25tf
    
    Vinci34tf=pow(ArcusDiff34tf,alpha) * Gini34tf
    Vinci35tf=pow(ArcusDiff35tf,alpha) * Gini35tf
   
    Vinci45tf=pow(ArcusDiff45tf,alpha) * Gini45tf
    
    VinciSumtf=(Vinci12tf+Vinci13tf+Vinci14tf+Vinci15tf+
                  Vinci23tf+Vinci24tf+Vinci25tf+
                  Vinci34tf+Vinci35tf+
                  Vinci45tf
                 )
    
    VinciMittelValuetf=VinciSumtf/10
    preVincitf=VinciMittelValuetf/pow(90,alpha)
    Vincitf=(preVincitf-(preVincitf*(1/5)))
    
    Vincid=100-((100/Vinci)*Vincitf)
    
    ###########################################################################################
    #Vinci tf2
    
    Vinci12tf2=pow(ArcusDiff12tf2,alpha) * Gini12tf2
    Vinci13tf2=pow(ArcusDiff13tf2,alpha) * Gini13tf2
    Vinci14tf2=pow(ArcusDiff14tf2,alpha) * Gini14tf2
    Vinci15tf2=pow(ArcusDiff15tf2,alpha) * Gini15tf2
   
    Vinci23tf2=pow(ArcusDiff23tf2,alpha) * Gini23tf2
    Vinci24tf2=pow(ArcusDiff24tf2,alpha) * Gini24tf2
    Vinci25tf2=pow(ArcusDiff25tf2,alpha) * Gini25tf2
    
    Vinci34tf2=pow(ArcusDiff34tf2,alpha) * Gini34tf2
    Vinci35tf2=pow(ArcusDiff35tf2,alpha) * Gini35tf2
    
    Vinci45tf2=pow(ArcusDiff45tf2,alpha) * Gini45tf2
   
    VinciSumtf2=(Vinci12tf2+Vinci13tf2+Vinci14tf2+Vinci15tf2+
                   Vinci23tf2+Vinci24tf2+Vinci25tf2+
                   Vinci34tf2+Vinci35tf2+
                   Vinci45tf2
                 )
    
    VinciMittelValuetf2=VinciSumtf2/10
    preVincitf2=VinciMittelValuetf2/pow(90,alpha)
    Vincitf2=(preVincitf2-(preVincitf2*(1/5)))
    
    Vincid2=100-((100/Vinci)*Vincitf2)
    
    ##############################################################################################
    
    percent1=round(g1*100,10)
    percent2=round(g2*100,10)
    percent3=round(g3*100,10)
    percent4=round(g4*100,10)
    percent5=round(g5*100,10)
      
    percent1tf=round(g1tf*100,10)
    percent2tf=round(g2tf*100,10)
    percent3tf=round(g3tf*100,10)
    percent4tf=round(g4tf*100,10)
    percent5tf=round(g5tf*100,10)
   
    percent1tf2=round(g1tf2*100,10)
    percent2tf2=round(g2tf2*100,10)
    percent3tf2=round(g3tf2*100,10)
    percent4tf2=round(g4tf2*100,10)
    percent5tf2=round(g5tf2*100,10)
    
    ##############################################################################################
    
    value1=a
    value2=b
    value3=c
    value4=d
    value5=e
   
    value1tf=atf
    value2tf=btf
    value3tf=ctf
    value4tf=dtf
    value5tf=etf
   
    value1tf2=atf2
    value2tf2=btf2
    value3tf2=ctf2
    value4tf2=dtf2
    value5tf2=etf2
   
    #############################################################################################
    if((Ginid>0.000000001)and(Difference1==Difference2)and(((TransferNr1<TransferNr2)and(Vincid<Vincid2))or
       ((TransferNr1>TransferNr2)and(Vincid>Vincid2)))and
       ((a!=b)and(a!=c)and(a!=d)and(a!=e)and
        (b!=c)and(b!=d)and(b!=e)and
        (c!=d)and(c!=e)and
        (d!=e))):
        DimTransferPrinciple="Violated"
    
    elif((Ginid>0.00000001)and(Difference1==Difference2)and(((TransferNr1<TransferNr2)and(Vincid>Vincid2))or
        ((TransferNr1>TransferNr2)and(Vincid<Vincid2)))and
        ((a!=b)and(a!=c)and(a!=d)and(a!=e)and
        (b!=c)and(b!=d)and(b!=e)and
        (c!=d)and(c!=e)and
        (d!=e))):
        DimTransferPrinciple="Satisfied"
        
   
    elif(abs(Ginid)<0.000000000001):
         DimTransferPrinciple="NoDifferenceGini"
            
    elif(Difference1!=Difference2):
         DimTransferPrinciple="DifferenceNotEqual" 
            
    elif((a==b)or(a==c)or(a==d)or(a==e)or
         (b==c)or(b==d)or(b==e)or
         (c==d)or(c==e)or
         (d==e)):
         DimTransferPrinciple="SameValues"
            
    elif(TransferNr1 == TransferNr2):
         DimTransferPrinciple="SameTransfer"
          
    else:
         DimTransferPrinciple="Other"   
     
    #################################################################################
    i+=1
   
    
    dataframe2.loc[len(dataframe2)] = (Quantile1,TransferNr1,TransferNr2,Quantile2,
                                       TransferSum1,TransferSum2,Mean,
                                       Vinci,Vincitf,Vincid,Vincitf2,
                                       Vincid2,alpha,DimTransferPrinciple,
                                       Gini,Ginitf,Ginitf2,Ginid,Ginid2,
                                       percent1,percent2,percent3,percent4,percent5,
                                       percent1tf,percent2tf,percent3tf,percent4tf,percent5tf,
                                       percent1tf2,percent2tf2,percent3tf2,percent4tf2,percent5tf2,
                                       value1,value2,value3,value4,value5,
                                       value1tf,value2tf,value3tf,value4tf,value5tf,
                                       value1tf2,value2tf2,value3tf2,value4tf2,value5tf2
                                      )                                               
                                                                          
  
        
```
